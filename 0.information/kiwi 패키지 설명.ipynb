### 불용어 방지
---

```python
from kiwipiepy.utils import Stopwords
stopwords = Stopwords()
kiwi.tokenize("카운더 보시는 할아버지만 불친절하지 음식은 맛있는 편이고 점심에 사람들 바글바글해서 대기시간이 좀 긴 편.", stopwords=stopwords)
```

---
###  토큰의 정보 
---

```python
tokens = kiwi.tokenize("각 토큰은 여러 정보를 담고 있습니다.")
tokens[0]
```

```python
print(tokens[0].form, '\n') # 형태소의 형태 정보

print(tokens[0].tag, '\n') # 형태소의 품사 정보

print(tokens[0].start, '\n') # 형태소의 문자 단위 시작점

print(tokens[0].end, '\n') # 형태소의 문자 단위 끝점

print(tokens[0].word_position, '\n') # 현 문장에서의 어절 번호

print(tokens[0].sent_position, '\n') # 형태소가 속한 문장 번호

print(tokens[0].line_number) # 형태소가 속한 줄의 번호
```

### 문장 분리 
---

```python
kiwi.split_into_sents("도가니탕 17000원 비싼데 가성비 너무 안좋네요.  포장하면 양이 엄청 적어요. 매장에서 먹었을때랑 심하게 차이나요. 파는 씻지도 않은거 흙 묻어있고 참..")
```

```python
kiwi.split_into_sents('친절하구 맛잇어요 파김치 별미... 헿')
```

```python
kiwi.split_into_sents("맛은 문제가 없지만 가성비가 너무 나쁘다. 10,000원 짜리 돈수백을 시키면 다른 집 5 ~ 6천원 보쌈정식보다 별로인 상이 나온다. 고기 양도 적고 마늘, 양파, 쌈장, 야채는 전혀 없다. 재방문의사는 물론 없다.")
```
