```python
def scroll():
    # 스크롤 전 높이
    before_h = browser.execute_script("return window.scrollY") # 현재 스크롤 된 높이를 계산

    # 무한 반복문 사용
    while True:
        # 맨 아래로 스크롤을 내린다.
        browser.find_element_by_css_selector("body").send_keys(Keys.END) # end키를 통해 스크롤을 맨 아래로 내림 
        
        # 스크롤 사이 페이지 로딩 시간 추가
        time.sleep(2)
        
        # 스크롤 후 높이 확인 
        after_h = browser.execute_script("return window.scrollY")
        
        if after_h == before_h: # 스크롤 높이가 마지막과 같다면 break
            break
        
        before_h = after_h # after의 높이를 before로 갱신
```

```python
import csv 
```

```python
# 파일 생성 
f = open(r"/Users/keonkookkim/Desktop/KKK/Selenium/data.csv", 'w', encoding="UTF-8", newline="") # 파일 입출력 기능 

csvWirter = csv.writer(f) # wirter에 파일(f) 넘겨주기

```

```python
# 상품 정보 div
items = browser.find_elements_by_css_selector(".basicList_info_area__17Xyo") # 리스트로 받기 위해서 element's'의 s를 붙여 줘야 함

for item in items:
    name = item.find_element_by_css_selector(".basicList_title__3P9Q7").text # text만 추출 
    price = item.find_element_by_css_selector(".price_num__2WUXn").text
    link = item.find_element_by_css_selector(".basicList_title__3P9Q7 > a ").get_attribute('href') # href 속성값 추출 
                                                    # 아래에 있는 a 태그 선택                           # href = link
    print(name, price, link)
    # 데이터 쓰기
    csvWirter.writerow([name, price, link]) # 행 추가

# 파일 닫기
f.close()

# 가격 정보가 없다면 try except 예외처리 
# try :
#     price = ~~~~
# except:
#     price = '판매중단'
```
